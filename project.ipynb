## 17. Build chains of trips. Two trips are consecutive in a chain if (a) they have the same VendorID, (b) the pickup location of the second trip is also the dropoff location of the first trip, (c) the pickup time of the second trip is after the dropoff time of the first trip, and (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip.
Hint: Add a column chain to the dataset. A chain can have more than two trips.


```python
original_df = df.copy()
```

```python
df = original_df.sample(n = 100000, replace = False)
```

```python
df['chain'] = pd.NA
df = df.sort_values(['VendorID', 'tpep_pickup_datetime']).reset_index(drop=True)

chain_id = 0

for vendor_id in df['VendorID'].unique():
        
    vendor_df = df[df['VendorID'] == vendor_id]
    
    # Pickup Location to Indexes Dictionary
    pickup_dict = {}
    for idx, row in vendor_df.iterrows():
        pickup_loc = row['PULocationID']
        if pickup_loc not in pickup_dict:
            pickup_dict[pickup_loc] = []
        pickup_dict[pickup_loc].append((row['tpep_pickup_datetime'], idx))
    
    # Sort Lists in the Dictionary
    for loc in pickup_dict:
        pickup_dict[loc].sort()
    
    processed = set()
    
    for idx in vendor_df.index:
        
        # Check if already processed
        if idx in processed:
            continue
        
        # To track the Chain Progress
        current_idx = idx
       
        # Start Chain
        current_chain = [idx]
        processed.add(idx)
        
        while True:
            current_trip = df.loc[current_idx]
            
            # Location Check
            location = current_trip['DOLocationID']          
            if location not in pickup_dict:
                break
            
            # Get List of Candidates
            candidates = pickup_dict[location]
            
            best_candidate = None
            
            for pickup_time, candidate_idx in candidates:
                # Check if Processed
                if candidate_idx in processed:
                    continue
                
                # Check First Condition of Time Matching
                if pickup_time <= current_trip['tpep_dropoff_datetime']:
                    continue
                
                # Check Second Condition of Time Matching
                if pickup_time <= current_trip['tpep_dropoff_datetime'] + pd.Timedelta('2min'):
                    best_candidate = candidate_idx
                break # If matches, it is the best, if doesnt match, no else will
            
            # No Best Candidate
            if best_candidate is None:
                break
            
            # Best Candidate    
            current_chain.append(best_candidate)
            processed.add(best_candidate)
            current_idx = best_candidate
        
        # Assegna chain_id
        if len(current_chain) > 1:
            df.loc[current_chain, 'chain'] = chain_id
            chain_id += 1
        else:
            df.loc[current_chain[0], 'chain'] = 'N'

df['chain'].value_counts()
```
